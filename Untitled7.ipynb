```python
import numpy as np
import pandas as pd
import scipy.stats as stats
from sklearn.model_selection import train_test_split
from sklearn.linear_model import Lasso, Ridge
from sklearn.preprocessing import StandardScaler

# 1. 數據模擬：零售損失數據
np.random.seed(42)
n_records = 1000

# 模擬不同倉庫的損失額
warehouses = ['A', 'B', 'C']
df = pd.DataFrame({
'Warehouse': np.random.choice(warehouses, n_records),
'Distance': np.random.uniform(10, 500, n_records), # 運輸距離
'Package_Weight': np.random.uniform(1, 50, n_records),
'Temperature': np.random.uniform(-5, 35, n_records),
'Loss_Amount': np.random.exponential(200, n_records) # 損失呈指數分佈 (常見於精算)
})

# 2. ANOVA 測試：不同倉庫的損失是否有顯著差異？
groups = [df[df['Warehouse'] == w]['Loss_Amount'] for w in warehouses]
f_stat, p_val = stats.f_oneway(*groups)
print(f"ANOVA F-statistic: {f_stat:.4f}, p-value: {p_val:.4f}")

# 3. Loss Elimination Ratio (LER) 計算
# 假設我們設定一個免賠額 (Deductible) d = 100
d = 100
total_loss = df['Loss_Amount'].sum()
eliminated_loss = df['Loss_Amount'].clip(upper=d).sum()
ler = eliminated_loss / total_loss
print(f"Loss Elimination Ratio (Deductible {d}): {ler:.2%}")

# 4. Lasso & Ridge Regression
# 預處理
X = pd.get_dummies(df.drop(columns=['Loss_Amount']), drop_first=True)
y = df['Loss_Amount']
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Lasso (會將無用特徵設為0)
lasso = Lasso(alpha=0.1)
lasso.fit(X_scaled, y)

# Ridge (平衡所有特徵權重)
ridge = Ridge(alpha=1.0)
ridge.fit(X_scaled, y)

print("\nLasso Coefficients:", dict(zip(X.columns, lasso.coef_)))
print("Ridge Coefficients:", dict(zip(X.columns, ridge.coef_)))
```
